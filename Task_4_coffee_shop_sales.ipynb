# ** ASSUMING COST PRICE TO 40% OF THE UNIT PRICE**

```python

# 2️⃣ Apply same 40% margin to all products
margin = 0.40
df['Cost_Price'] = df['unit_price'] * (1 - margin)

# 3️⃣ Check the result
print(df[['product_type', 'unit_price', 'Cost_Price']].tail())
```

# **# 2️⃣ Calculate Profit per transaction**

```python

df['Profit'] = (df['unit_price'] - df['Cost_Price']) * df['transaction_qty']

# 3️⃣ Preview the result
print(df[['product_type', 'unit_price', 'Cost_Price', 'transaction_qty', 'Profit']].tail())

# 4️⃣ Total profit for dataset
total_profit = df['Profit'].sum()
print(f"\n💰 Total Estimated Profit: ${total_profit:.2f}")
```

```python

# --- STEP 1: Feature Engineering ---
df['is_weekend'] = df['Day Name'].isin(['Saturday', 'Sunday']).astype(int)

def time_of_day(hour):
    if 5 <= hour < 12:
        return 'Morning'
    elif 12 <= hour < 17:
        return 'Afternoon'
    else:
        return 'Evening'

df['time_of_day'] = df['Hour'].apply(time_of_day)

# --- STEP 2: Grouping for Profit Analysis ---
# Profit by Product Category
profit_by_category = df.groupby('product_category')['Profit'].sum().sort_values(ascending=False)
print("\n💰 Profit by Product Category:\n", profit_by_category)

# Profit by Product Type
profit_by_type = df.groupby('product_type')['Profit'].sum().sort_values(ascending=False)
print("\n💰 Profit by Product Type:\n", profit_by_type)

# Profit by Size
profit_by_size = df.groupby('Size')['Profit'].sum().sort_values(ascending=False)
print("\n💰 Profit by Size:\n", profit_by_size)

# Profit by Store Location
profit_by_location = df.groupby('store_location')['Profit'].sum().sort_values(ascending=False)
print("\n💰 Profit by Store Location:\n", profit_by_location)

# Profit by Day Name
profit_by_day = df.groupby('Day Name')['Profit'].sum().reindex(
    ["Monday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday"]
)
print("\n💰 Profit by Day of Week:\n", profit_by_day)

# Profit by Hour
profit_by_hour = df.groupby('Hour')['Profit'].sum().sort_values(ascending=False)
print("\n💰 Profit by Hour:\n", profit_by_hour)

# --- STEP 3: Feature Importance (Regression) ---
from sklearn.preprocessing import OneHotEncoder
from sklearn.linear_model import LinearRegression

# Select features for model
X = df[['store_location','product_category','product_type','Size','Day Name','Hour','is_weekend','time_of_day']]
y = df['Profit']

# One-hot encode categorical variables
X_encoded = pd.get_dummies(X, drop_first=True)

# Fit linear regression
model = LinearRegression()
model.fit(X_encoded, y)

# Create feature importance DataFrame
feature_importance = pd.DataFrame({
    'Feature': X_encoded.columns,
    'Importance': model.coef_
}).sort_values(by='Importance', ascending=False)

print("\n📊 Feature Importance for Profit Prediction:\n", feature_importance)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1️⃣ Profit by Product Category
plt.figure(figsize=(8,5))
sns.barplot(x=profit_by_category.index, y=profit_by_category.values, palette="viridis")
plt.title("💰 Profit by Product Category")
plt.ylabel("Total Profit ($)")
plt.xlabel("Product Category")
plt.xticks(rotation=45)
plt.show()

# 2️⃣ Profit by Product Type (Top 10)
plt.figure(figsize=(10,6))
top_types = profit_by_type.head(10)
sns.barplot(x=top_types.values, y=top_types.index, palette="coolwarm")
plt.title("💰 Top 10 Profitable Product Types")
plt.xlabel("Total Profit ($)")
plt.ylabel("Product Type")
plt.show()

# 3️⃣ Profit by Size
plt.figure(figsize=(6,4))
sns.barplot(x=profit_by_size.index, y=profit_by_size.values, palette="magma")
plt.title("💰 Profit by Size")
plt.ylabel("Total Profit ($)")
plt.show()

# 4️⃣ Profit by Store Location
plt.figure(figsize=(8,5))
sns.barplot(x=profit_by_location.index, y=profit_by_location.values, palette="cubehelix")
plt.title("💰 Profit by Store Location")
plt.ylabel("Total Profit ($)")
plt.show()

# 5️⃣ Profit Heatmap by Day & Hour
heatmap_profit = df.groupby(['Day Name', 'Hour'])['Profit'].sum().unstack(fill_value=0)
heatmap_profit = heatmap_profit.reindex(["Monday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday"])

plt.figure(figsize=(12,6))
sns.heatmap(heatmap_profit, cmap="YlGnBu", annot=True, fmt=".0f")
plt.title("💰 Profit Heatmap by Day & Hour")
plt.ylabel("Day of Week")
plt.xlabel("Hour of Day")
plt.show()

#  Profit by Time of Day
plt.figure(figsize=(6,4))
sns.barplot(x=df.groupby('Month')['Profit'].sum().index,
            y=df.groupby('Month')['Profit'].sum().values,
            palette="plasma")
plt.title("💰 Profit By Month")
plt.ylabel("Total Profit ($)")
plt.show()

# 6️⃣ Profit by Time of Day
plt.figure(figsize=(6,4))
sns.barplot(x=df.groupby('time_of_day')['Profit'].sum().index,
            y=df.groupby('time_of_day')['Profit'].sum().values,
            palette="plasma")
plt.title("💰 Profit by Time of Day")
plt.ylabel("Total Profit ($)")
plt.show()
```

# **1️⃣ Prepare the Time Series Data**
We first combine your date elements into a proper datetime column and sum sales/profit per day.

```python
import pandas as pd

# Ensure date is in datetime format
df['date'] = pd.to_datetime(df['date'])

#Aggregate for getting monthly forecasts:

monthly_df = df.groupby(pd.Grouper(key='date', freq='M')).agg({
    'Total_Bill': 'sum',
    'Profit': 'sum'
}).reset_index()

#3️⃣ Sort chronologically

monthly_df = monthly_df.sort_values('date')

#4️⃣ Rename for Prophet

prophet_df = monthly_df.rename(columns={'date': 'ds', 'Profit': 'y'})[['ds', 'y']]

prophet_df



```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
plt.plot(df['date'], df['Profit'], marker='o')
plt.title("Daily Profit Trend")
plt.xlabel("Date")
plt.ylabel("Profit ($)")
plt.grid(True)
plt.show()
```
